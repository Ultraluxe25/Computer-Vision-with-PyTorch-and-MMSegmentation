**Задачи:**

1.	Подготовить набор данных для N (5-10) классов, для каждого из классов подготовить минимум 100 примеров изображений
2.	Организовать набор данных в виде:
    - Директория, содержащая все изображения
    - Файл images.json, который содержит список объектов с полями - идентификатор, имя файла и класс изображения
    - Файл train.json, содержащий список с идентификаторами изображений из обучающей выборки
    - Файл validation.json, содержащий список с идентификаторами изображений из контрольной выборки
3.	В проект обучения нейронной сети для классификации изображений FashionMnist (семинар №1) внести изменения:
    - Реализовать пользовательский датасет для структуры данных, описанных в пункте 2 https://pytorch.org/tutorials/beginner/basics/data_tutorial.html 
4.	Обучить нейронную сеть пользовательской архитектуры для классификации изображений 
5.	Исследовать зависимость характеристик модели от гиперпараметров обучения - темпа обучения, выбора оптимизатора, параметров оптимизатора
6.	Дообучить предобученную модель resnet18 https://pytorch.org/tutorials/beginner/transfer_learning_tutorial.html#finetuning-the-convnet для решения задачи классификации и сравнить результаты с 3


### ResNet18 дала хуре результат чем пользовательская нейронка, т.к. ResNet18 переобучилась на FashionMNIST

### Очень важно подобрать хорошую валидационную выборку, чтобы не переобучиться!

### Лучше не обрезать исходное изображение, чтобы можно было сделать crop-аугментацию

### Для своего пользовательского датасета достаточно взять 5-10 классов и 100+ изображений для каждого

### Создать мета-файл JSON, в котором будет вся информация по датасету. Классы должны быть НЕ ПЕРЕСЕКАЮЩИМИСЯ! (15-20 минуты видео)
